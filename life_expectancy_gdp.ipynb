#### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# load the data
df = pd.read_csv('all_data.csv')
df.rename(columns={"Life expectancy at birth (years)": "LEABY"}, inplace=True)
sns.set()

df
```

**Variables/Columns**
- Country - nation
- Year - the year for the observation
- Life expectancy at birth (years) - life expectancy value in years
- GDP - Gross Domestic Product in U.S. dollars

## Project goals:
1. Has life expectancy increased over time in the six nations?
2. Has GDP increased over time in the six nations?
3. Is there a correlation between GDP and life expectancy of a country?
4. What is the average life expectancy in these nations?
5. What is the distribution of that life expectancy?

## 1. Has life expectancy increased over time in the six nations?


```python
# # create a line plot to see change in life expectancy over time
g = sns.FacetGrid(df, col="Country", col_wrap=3, height=4)
g.map(plt.plot, "Year", "LEABY", color='m').add_legend()

plt.show()
plt.clf()
```

**There is a positive trend upwards for life expectancy for the 6 countries**

## 2. Has GDP increased over time in the six nations?


```python
# plot gdp and years
    
g2 = sns.FacetGrid(df, col="Country", col_wrap=3, height=4)
g2.map(plt.plot, "Year", "GDP").add_legend()

plt.show()
plt.clf()
```

**Each country has seen a rise in GDP, some larger than others, but all in an uppward trend.**

## 3. Is there a correlation between GDP and life expectancy of a country?


```python
countries = df.Country.unique()

for country in countries:
    sns.scatterplot(x=df.GDP[df.Country == country], y=df.LEABY[df.Country == country])
    # title
    plt.title(country + ' correlation between GDP and LEABY')
    plt.xlabel('GDP (in trillions of USD)')
    plt.ylabel('LEABY')
    plt.show()
    plt.clf()
```

**According to this data as a countries life expectancy rises so does thei GDP**

## 4. What is the average life expectancy in these nations?


```python
ax = plt.subplot()
sns.barplot(x=df.Country, y=df.LEABY, data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=70)

plt.show()
plt.clf()

for country in countries:
    average_country_LEABY = round(np.mean(df.LEABY[df.Country == country]), 2)
    print(country + "'s average life expectancy: ", average_country_LEABY)
```
